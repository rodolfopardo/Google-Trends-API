```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.tail()
```

```python
format_dict = {'date':'{:%d-%m-%Y}'} #Simplified format dictionary with values that do make sense for our data
data.tail(30).style.format(format_dict).highlight_max(color='red').highlight_min(color='blue')
```

```python
data.tail(60).style.format(format_dict).background_gradient(subset=['Coronavirus', 'China', 'sintomas', 'Youtube'], cmap='BuGn')
```

```python
data.tail(60).style.format(format_dict).bar(color='green', subset=['Coronavirus', 'sintomas', 'China', 'Youtube'])
```

```python
data.tail(60).style.format(format_dict).background_gradient(subset=['Coronavirus', 'sintomas', 'China', 'Youtube'], cmap='BuGn').highlight_min(color='yellow')
```

```python
corr = data.corr()
```

```python
corr.style.background_gradient(cmap='coolwarm')
```

```python
sns.lineplot(x = data['date'], y = data['Coronavirus'], label = 'Coronavirus')
sns.lineplot(x = data['date'], y = data['sintomas'], label = 'Sintomas')
sns.lineplot(x = data['date'], y = data['China'], label = 'China')
sns.lineplot(x = data['date'], y = data['Youtube'], label = 'Youtube');
```

```python
#Graficando correlacion 

plt.scatter(data['Coronavirus'], data['sintomas'])
```
