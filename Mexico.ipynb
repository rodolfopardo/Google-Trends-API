```python
import pytrends
from pytrends.request import TrendReq
import pandas as pd
import time
import datetime
from datetime import datetime, date, time
from pandas.io.json import json_normalize
```

```python
pytrend = TrendReq()
pytrend.build_payload(kw_list=['Yogur'], timeframe='today 12-m', geo = 'MX')
```

```python
interest_over_time_df = pytrend.interest_over_time() 
print(interest_over_time_df.head())
```

```python
# Let's draw
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=False)
dx = interest_over_time_df.plot.line(figsize = (13,6), title = "Inters en el tiempo", color = "blue", linewidth = 2)
dx.set_xlabel('Fecha')
dx.set_ylabel('Indexaciones en bsqueda')
dx.tick_params(axis='both', which='major', labelsize=16)
```
